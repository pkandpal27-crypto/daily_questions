```python
def r_d(arr):
    l = len(arr)
    if l == 0:
        return 0
    
    i = 0  # index of last unique element
    for j in range(1, l):
        if arr[j] != arr[i]:
            i += 1
            arr[i] = arr[j]
    return i + 1  # length of unique elements

# Testing
arr1 = [1,1,2,2,3,3,4,4,5,6]
arr2 = [1,2,3,4,5,6]
arr3 = [5,5,5,5,5,5]

print(r_d(arr1))  # Output: 6
print(arr1[:r_d(arr1)])  # Unique elements: [1,2,3,4,5,6]

print(r_d(arr2))  # Output: 6
print(arr2[:r_d(arr2)])  # Unique elements: [1,2,3,4,5,6]

print(r_d(arr3))  # Output: 1
print(arr3[:r_d(arr3)])  # Unique elements: [5]
```
