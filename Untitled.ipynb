```python
#find the factors of a number;
def factors(n):
    limit=int(n**0.5)
    fac=[]
    for i in range(1,limit):
        if n%i==0:
            fac.append(i)
            if n//i!=i:
                fac.append(n//i)
    fac.sort()
    return fac

print(factors(45))
```

```python
#frequency of letters in a string
def freq(st):
    fd={}
    for i in st:
        if i not in fd:
            fd[i]=1
        else:
            fd[i]=fd[i]+1
        
    return fd
st="ashutosh kandpal is what you'd call the god of coding?"
print(freq(st))
```

```python
#recursion using tail recursion
def printn(n):
    if n==0:
        return
    printn(n-1)
    print(n, end= " ")


printn(10)
    
```

```python
#check palindrome using recursion
def ispalindrome(st,s,e):
    if s>=e:
        return True
    if st[s]==st[e]:
        return ispalindrome(st,s+1,e-1)
    else:
        return False



word=input("enter some string: ")
length=len(word)
ispalindrome(word,0,length-1)
```
